# **1. Import Library**

# **2. Read Dataset**

```python
import pandas as pd
from sklearn import datasets

# Memuat dataset Breast Cancer dari scikit-learn
cancer = datasets.load_breast_cancer()

# input utk Machine Learning
x = cancer.data
# outlet yg diinginkan
y = cancer.target

# Mengonversi data fitur dan target menjadi DataFrame
df_x = pd.DataFrame(x, columns=cancer.feature_names)
df_y = pd.Series(y, name='target')
```

```python
# Gabungkan fitur dan target dalam satu DataFrame
df_cancer = pd.concat([df_x, df_y], axis=1)

# Menampilkan 15 data teratas
df_cancer.head(15)
```

# **3. Split Data**

```python
from sklearn.model_selection import train_test_split

# Membagi data menjadi LATIH(TRAIN) dan UJI(TEST)
x_latih, x_uji, y_latih, y_uji = train_test_split(df_x, df_y, test_size=0.2, random_state=42)
```

# **4. Train the Model**

```python
# @title Model 1 Logistic Regression
from sklearn.linear_model import LogisticRegression

# Membuat dan melatih model Logical Regression
model_1 = LogisticRegression(random_state=42)
model_1.fit(x_latih, y_latih)
```

```python
# @title Model 2 Random Forest
from sklearn.ensemble import RandomForestClassifier

# Membuat dan melatih model Logical Regression
model_2 = RandomForestClassifier(random_state=42)
model_2.fit(x_latih, y_latih)
```

# **5. Predict & Evaluate**

```python
# @title Matrix Model 1 - Logistic Regression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Memprediksi dan mengevaluasi
y_predict_LR = model_1.predict(x_uji)
accuracy = accuracy_score(y_uji, y_predict_LR)

print("Laporan Klasifikasi")
print(f"Akurasi Logistic Regression = {accuracy * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix/CM
CM = confusion_matrix(y_uji, y_predict_LR)

# Plot Confusion Matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(CM, annot=True, fmt="d", cmap="Reds",
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confusion Matrix Logistic Regression")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_uji, y_predict_LR, target_names=cancer.target_names))
```

```python
# @title Matrix Model 2 - Random Forest
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Memprediksi dan mengevaluasi
y_predict_RF = model_2.predict(x_uji)
accuracy = accuracy_score(y_uji, y_predict_RF)

print("Laporan Klasifikasi")
print(f"Akurasi Random Forest = {accuracy * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix/CM
CM = confusion_matrix(y_uji, y_predict_RF)

# Plot Confusion Matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(CM, annot=True, fmt="d", cmap="Blues",
            xticklabels=cancer.target_names, yticklabels=cancer.target_names)
plt.xlabel("Predict")
plt.ylabel("Actual")
plt.title("Confusion Matrix Random Forest")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(classification_report(y_uji, y_predict_RF, target_names=cancer.target_names))
```
